```python
%matplotlib inline
import matplotlib
import cmath
import math
from matplotlib.ticker import MultipleLocator
import matplotlib.pyplot as plt
import numpy as np
import decimal

#Define complex number
ComplexNo_RealPart=-4
ComplexNo_ImagPart=3
z=complex(ComplexNo_RealPart,ComplexNo_ImagPart)
ComplexNo=cmath.polar(z)
z_module=ComplexNo[0]
z_arg=ComplexNo[1]
z_module_str=str(z_module)
z_arg_str=str(z_arg)

# define X range adjusted to complex number by 50%
NoPercAdjust=0.5 #50%
max_number=max(abs(ComplexNo_RealPart),abs(ComplexNo_ImagPart))
max_scale=int(math.ceil(max_number*(1+NoPercAdjust)))

grXmin=-max_scale#Graph_Range_X_Min=0
grXmax=max_scale#Graph_Range_X_Max=10
grXdivisions=1#Graph_Range_X_Division=1
Graph_Range_X=np.arange(grXmin,grXmax+1,grXdivisions)
# define Y range
grYmin=-max_scale#Graph_Range_Y_Min=0
grYmax=max_scale#Graph_Range_Y_Max=10
grYdivisions=1#Graph_Range_Y_Division=1
Graph_Range_Y=np.arange(grYmin,grYmax+1,grXdivisions)

#Define data to plot X axis
Xaxis_X_Data=np.linspace(grXmin,grXmax,num=(grXmax-grXmin+1),endpoint=True)
Xaxis_Y_Data=np.full((grXmax-grXmin+1), 0)

#Define data to plot Y axis
Yaxis_X_Data=np.full((grYmax-grYmin+1), 0)
Yaxis_Y_Data=Graph_Range_Y

# Plots a X axis
x1 = Xaxis_X_Data
y1 = Xaxis_Y_Data
#Plots a Y axis
x2=Yaxis_X_Data
y2=Yaxis_Y_Data

#Define figure to plot size, dpi
fig=plt.figure(figsize=[2.5,2.5], dpi=200)
ax=fig.add_subplot(111)

#Define Grid lines spacing
spacing = 1#max_scale/6 # This can be your user specified spacing. 
minorLocator = MultipleLocator(spacing)
ax.yaxis.set_minor_locator(minorLocator)
ax.xaxis.set_minor_locator(minorLocator)

#Plot data
ax.plot(x1,y1,x2,y2) #plot 2 graps on same area

#Plot One Complex Number cartesian
ax.plot([ComplexNo_RealPart], [ComplexNo_ImagPart], 'o') # Defines the dot symbol on the plane
ax.annotate('', 
            xy=(ComplexNo_RealPart,ComplexNo_ImagPart), 
            xytext=(0, 0),
            arrowprops=dict(facecolor='green',
                           width=2, #Defines arrow width
                           headwidth=8))#Defines arrow head width

#add the X axis arrow
ComNoLabel="X"
ax.text(grXmax, 0, ComNoLabel, fontsize=5)
ax.annotate('', 
            xy=(grXmax,0), 
            xytext=(grXmin, 0),
            arrowprops=dict(facecolor='blue',
                           width=2, #Defines arrow width
                           headwidth=8))#Defines arrow head width

#add the Y axis arrow
ComNoLabel="Y"
ax.text(0,grYmax,ComNoLabel, fontsize=5)
ax.annotate('', 
            xy=(0,grYmax), 
            xytext=(0,grYmin),
            arrowprops=dict(facecolor='orange',
                           width=2, #Defines arrow width
                           headwidth=8))#Defines arrow head width

#Plot labels, title and grid details
ax.set(xlabel='Real - X', 
       ylabel='Imaginary - Y',
       title='ARGAND DIAGRAM')
ax.grid(which='minor',axis='both',color='black', linestyle='-.', linewidth=0.3)# shows minor grid

#PLot complex number values
ComNoLabelCartesian="("+ str(ComplexNo_RealPart)+","+str(ComplexNo_ImagPart)+")"
ax.text(ComplexNo_RealPart, ComplexNo_ImagPart+0.5, ComNoLabelCartesian, fontsize=5,color='blue') # Add label to complex number cartesian coordinates
ComNoLabelPolar="(|z|= "+ str(z_module_str)+",arg = "+str(z_arg_str)+")"
ax.text(ComplexNo_RealPart, ComplexNo_ImagPart, ComNoLabelPolar, fontsize=5,color='red') # Add label to complex number polar coordinates

#Adding angle arrow


#Show graphics
plt.show()
```
